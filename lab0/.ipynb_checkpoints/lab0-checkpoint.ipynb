```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
K = 7
N = 19
M = (4 % 9)* 10
```

```python
# Read the first K + N data
data = pd.read_csv("Top_100_Tennis_Players-2007_Men.csv", nrows= K + N)

# delete % and print 
data['Winning Percentage'] = list(map(lambda x: x[:-1], data['Winning Percentage'].values))
data['Winning Percentage'] = [float(x) for x in data['Winning Percentage'].values]
data
```

```python
# 3.a

count_of_player = data["Country"]
count = 0
country = "Sweden"

# count players by country
for item in range(0, len(count_of_player)):
    if count_of_player[item].replace(' ', '') == country:
        count += 1

print("Count of player into {} = {}".format(country, count))
```

```python
# 3.b
sum_of_money = 0
count_of_player = 0
average = 0
# average value for players between M and M + 10
for item in range(0, len(data)):
    if M <= float(data["Winning Percentage"][item]) <= M + 10:
        sum_of_money += int(data.at[item, "Career Earnings"].replace('$',''))
        count_of_player += 1
    
if count_of_player != 0:
    average = sum_of_money / count_of_player
    print("Average money for players between {} and {} = {}".format(M, M + 10, average))
else:
    print("division by zero")
```

```python
# 4.a
# average rank players by country
rank_by_country = data.groupby("Country")["Rank"].mean()
# 4.b
# average winning % by country
winning_by_country = data.groupby("Country")["Winning Percentage"].mean()

df = rank_by_country.to_frame().join(winning_by_country)
# draw graph
df.plot(figsize=(10, 5),kind = 'bar')
```
