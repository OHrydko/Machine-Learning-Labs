```python
import pandas as pd


df = pd.read_csv('https://raw.githubusercontent.com/OHrydko/Machine-Learning-Labs/master/lab3/MP-07-Hrydko.csv', 
                 header=None, names=['x0', 'x1', 'x2', 'x3', 'x4', 'x5', 'x6','x7', 'x8', 'x9', 'x10', 'x11', 'y'], sep=';',engine='python')
y = df['y'].values
X = df[df.columns[:-1]].values

df.head()
```

```python
y_train = y[:200]
X_train = X[:200]

y_val = y[200:]
X_val = X[200:]
```

```python
N = len(X_train)
dim = len(X_train[0]) 
attrs = [[] for _ in range(dim)] 
output_dom = {}
data = [] 
    
def calculate(X, y):
    for i in range(len(X)):
        for j in range(dim):
            # унікальні значення x по стовпчиках
            if not X[i][j] in attrs[j]:
                attrs[j].append(X[i][j])
        # кількість edible та poisonous
        if not y[i] in output_dom.keys():
            output_dom[y[i]] = 1
        else:
            output_dom[y[i]] += 1
        data.append([X[i], y[i]])

def classify(row):

    solve = None 
    max_arg = -1 
    for y in output_dom.keys():
        prob = output_dom[y]/N # P(y)

        for i in range(dim):
            cases = []
            for x in data:
                if x[0][i] == row[i] and x[1] == y:
                    cases.append(x)
            n = len(cases)
            prob *= n/N # P *= P(Xi = xi)
            
        if prob > max_arg:
            max_arg = prob
            solve = y
    return solve
```

```python
calculate(X_train, y_train)
total_cases = len(y_val)

right = 0
wrong = 0

for i in range(total_cases):
    predict = classify(X_val[i])
#     print(y_val[i] + ' --------------- ' + predict)
    if y_val[i] == predict:
        right += 1
    else:
        wrong += 1

print('RIGHT:', right)
print('WRONG:', wrong)
print('ACCURACY {} %:'.format(100*right/total_cases))
```
